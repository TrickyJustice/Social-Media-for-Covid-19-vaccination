```python
for i in range(1,11):
  sum=0;
  for j in range (581):
    if(df1['category'][j]==i):
      a = df1['tweet'][j]
      c = sid.polarity_scores(a)
      sum = sum + c['compound']
  data.append(sum)
for i in range(1,11):
    sum=0;
    for j in range (532):
      if(df4['category'][j]==i):
        a = df4['tweet'][j]
        c = sid.polarity_scores(a)
        sum = sum + c['compound']
    datad.append(sum)
```
